```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

jee_marks = ctrl.Antecedent(np.arange(0, 301, 1), 'jee_marks')
school_marks = ctrl.Antecedent(np.arange(0, 101, 1), 'school_marks')
admission = ctrl.Consequent(np.arange(0, 11, 1), 'admission')

jee_marks["low"]=fuzz.trapmf(jee_marks.universe,[0,0,80,100])
jee_marks["medium"]=fuzz.trapmf(jee_marks.universe,[60,100,140,160])
jee_marks["high"]=fuzz.trapmf(jee_marks.universe,[150,200,250,260])
jee_marks["very_high"]=fuzz.trapmf(jee_marks.universe,[200,250,300,300])

school_marks["bad"]=fuzz.trapmf(school_marks.universe,[0,0,20,40])
school_marks["fair"]=fuzz.trapmf(school_marks.universe,[20,50,60,80])
school_marks["excellent"]=fuzz.trapmf(school_marks.universe,[60,80,100,100])

admission["very_low"]=fuzz.trimf(admission.universe,[0,0,3])
admission["low"]=fuzz.trimf(admission.universe,[0,3,6])
admission["medium"]=fuzz.trimf(admission.universe,[2,5,8])
admission["high"]=fuzz.trimf(admission.universe,[4,7,10])
admission["very_high"]=fuzz.trimf(admission.universe,[7,10,10])
```

```python
jee_marks.view()
```

```python
school_marks.view()
```

```python
admission.view()
```

```python
# Rule Application
rule1 = ctrl.Rule( jee_marks['low'] , admission['very_low'])
rule2 = ctrl.Rule( school_marks['bad'], admission['low'])
rule3 = ctrl.Rule( jee_marks['low'] | school_marks['bad'], admission['very_low'])
rule4 = ctrl.Rule( jee_marks['medium'] | school_marks['bad'], admission['low'])
rule5 = ctrl.Rule( jee_marks['high'] | school_marks['bad'], admission['medium'])
rule6 = ctrl.Rule( jee_marks['very_high'] | school_marks['bad'], admission['high'])
rule7 = ctrl.Rule( jee_marks['low'] | school_marks['fair'], admission['low'])
rule8 = ctrl.Rule( jee_marks['medium'] | school_marks['fair'], admission['medium'])
rule9 = ctrl.Rule( jee_marks['high'] | school_marks['fair'], admission['high'])
rule10 = ctrl.Rule( jee_marks['very_high'] | school_marks['fair'], admission['very_high'])
rule11 = ctrl.Rule( jee_marks['low'] | school_marks['excellent'], admission['medium'])
rule12 = ctrl.Rule( jee_marks['medium'] | school_marks['excellent'], admission['high'])
rule13 = ctrl.Rule( jee_marks['high'] | school_marks['excellent'], admission['very_high'])
rule14 = ctrl.Rule( jee_marks['very_high'] | school_marks['excellent'], admission['very_high'])
```

```python
admission_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, 
                                 rule8, rule9, rule10, rule11, rule12, rule13, rule14])
```

```python
admission_ = ctrl.ControlSystemSimulation(admission_ctrl)
```

```python
admission_.input["jee_marks"]=32
admission_.input["school_marks"]=3

admission_.compute()
```

```python
admission_.output["admission"]
```

```python
admission.view(sim=admission_)
```
